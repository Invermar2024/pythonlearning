# 04 - Files and Errors

This notebook is about saving data outside the program and handling mistakes without the program crashing immediately.

## Writing a text file

The `with open(...)` pattern closes the file for me when the block is done.

```python
notes = ["variables", "loops", "functions"]

with open("practice_notes.txt", "w", encoding="utf-8") as file:
    for note in notes:
        file.write(note + "\n")
```

## Reading the file back

I like using `.strip()` here because each line has a newline character at the end.

```python
with open("practice_notes.txt", "r", encoding="utf-8") as file:
    for line in file:
        print(line.strip())
```

## Try and except

`try` and `except` let me handle errors. I should not hide every error, but for expected problems this is useful.

```python
value = "abc"

try:
    number = int(value)
    print(number * 2)
except ValueError:
    print("That was not a number.")
```

## Mini exercise

Read numbers from a list of strings. Skip anything that cannot be converted.

```python
raw_values = ["10", "hello", "25", "", "7"]
numbers = []

for value in raw_values:
    try:
        numbers.append(int(value))
    except ValueError:
        pass

numbers
```
