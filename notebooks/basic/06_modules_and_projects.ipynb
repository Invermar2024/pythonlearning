## A tiny project shape

For a small program, I can start with three parts:

- input data
- logic that changes or checks the data
- output for the user

```python
task = {"title": "review functions", "done": False}

status = "done" if task["done"] else "pending"
print(f"Task: {task['title']}")
print(f"Status: {status}")
```
