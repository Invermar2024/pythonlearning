# 05 - Data Structures

This notebook compares the common containers I keep seeing in Python: lists, dictionaries, sets, and tuples.

## Lists

Lists are ordered and can change. I use them when I have several items in sequence.

```python
shopping = ["milk", "rice"]
shopping.append("eggs")
shopping
```

## Dictionaries

Dictionaries are good when each value needs a label.

```python
student = {"name": "Ana", "topic": "Python"}
f"{student['name']} is learning {student['topic']}"
```

## Sets

Sets keep unique values. They do not keep the same order idea as lists.

```python
file_types = ["csv", "txt", "csv", "json", "txt"]
unique_file_types = set(file_types)
unique_file_types
```

## Tuples

Tuples are ordered but not meant to change. Coordinates are a common example.

```python
point = (10, 20)
x, y = point
f"x={x}, y={y}"
```

## Mini exercise

Count how many times each topic appears in a list.

```python
topics = ["python", "sql", "python", "git", "python"]
counts = {}

for topic in topics:
    counts[topic] = counts.get(topic, 0) + 1

counts
```

My simple rule for now: list for ordered items, dictionary for labeled data, set for unique values, tuple for fixed pairs or groups.
