# Intermediate 01 - Comprehensions and Lambdas

This topic is about writing small transformations in a cleaner way. I do not want to overuse these, but they are useful when the logic is simple.

## List comprehensions

A list comprehension creates a new list from another iterable.

```python
numbers = [1, 2, 3, 4, 5]
squares = [number ** 2 for number in numbers]
squares
```

## Filtering

I can add an `if` at the end when I only want some values.

```python
values = [1, 2, 3, 4, 5, 6]
even_values = [value for value in values if value % 2 == 0]
even_values
```

## Dictionary comprehensions

This is similar, but I create key-value pairs.

```python
topics = ["python", "sql", "git"]
topic_lengths = {topic: len(topic) for topic in topics}
topic_lengths
```

## Lambda functions

A lambda is a small anonymous function. I usually prefer normal functions, but lambdas are common in sorting.

```python
students = [
    {"name": "Luis", "score": 85},
    {"name": "Ana", "score": 92},
    {"name": "Mia", "score": 78},
]

sorted(students, key=lambda student: student["score"], reverse=True)
```

Mini exercise: take a list of prices and create a new list with a 10 percent discount applied. Then filter only the prices below 100.

```python
prices = [100, 250, 75, 120]
discounted = [price * 0.9 for price in prices]
cheap_discounted = [price for price in discounted if price < 100]
cheap_discounted
```
