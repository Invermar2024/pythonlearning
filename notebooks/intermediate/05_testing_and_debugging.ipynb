# Intermediate 05 - Testing and Debugging

Testing is how I check that code still works after I change it. Debugging is how I investigate when it does not.

## Start with small functions

Small functions are easier to test than code that does many things at once.

```python
def add(a, b):
    return a + b


def is_even(number):
    return number % 2 == 0
```

## Assertions

`assert` checks that something is true. If it is false, Python raises an error.

```python
assert add(2, 3) == 5
assert is_even(4) is True
assert is_even(5) is False
```

## Testing edge cases

An edge case is an input that might behave differently from normal examples.

```python
def average(numbers):
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)


assert average([10, 20]) == 15
assert average([]) == 0
```

## Debugging habits

When something fails, I want to slow down and check:

- What input did I give the function?
- What did I expect?
- What did I actually get?
- Which line does the error mention?

This is less exciting than guessing, but it works better.

```python
values = [3, 6, 9]

for value in values:
    print(f"Checking value: {value}")
```
