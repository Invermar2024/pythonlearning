# Intermediate 03 - Modules, Packages, and Environments

This topic is about organizing code once a project starts growing beyond one file.

## Modules

A module is just a Python file that can be imported. For example, `math` is a module from the standard library.

````python
import math

math.sqrt(9)
````

## Import styles

There are different ways to import. I prefer the clearest option instead of the shortest one.

````python
from datetime import date

date(2026, 7, 9)
````

## Project layout idea

A small project can grow into a structure like this:

```text
my_project/
  README.md
  requirements.txt
  src/
    app.py
    helpers.py
  tests/
    test_helpers.py
```

I do not need this for every tiny script, but it helps once files start multiplying.

## Virtual environments

A virtual environment keeps project dependencies separate.

```bash
python3 -m venv .venv
source .venv/bin/activate
pip install requests
pip freeze > requirements.txt
```

Note to myself: do not commit `.venv/` to GitHub.

Mini exercise: move repeated functions from a project script into a separate `helpers.py` file, then import them from the main file.
