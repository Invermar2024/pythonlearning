This notebook practices common data formats. APIs often return JSON, while spreadsheets are often exported as CSV.

## JSON

JSON looks similar to Python dictionaries and lists, but it is a text format.

```python
import json

raw_json = '{"name": "Ana", "lessons_completed": 4}'
data = json.loads(raw_json)
data["name"]
```

```python
json.dumps({"topic": "Python", "level": "intermediate"})
```

## CSV

CSV is useful for table-shaped data.

```python
import csv
from io import StringIO

csv_text = "name,score\nAna,92\nLuis,85\n"
reader = csv.DictReader(StringIO(csv_text))
rows = list(reader)
rows
```

## API-style data

Instead of calling the internet here, I am using a local dictionary shaped like an API response. The parsing idea is the same.

```python
response = {
    "status": "ok",
    "items": [
        {"id": 1, "title": "learn functions", "done": True},
        {"id": 2, "title": "practice OOP", "done": False},
    ],
}

[item["title"] for item in response["items"]]
```

Mini exercise: count how many API items are done and how many are pending.

```python
summary = {"done": 0, "pending": 0}

for item in response["items"]:
    if item["done"]:
        summary["done"] += 1
    else:
        summary["pending"] += 1

summary
```
