# 03 - Functions

Functions let me give a name to a piece of code and reuse it. This is where code starts feeling more organized.

## A first function

A function starts with `def`. The code inside the function is indented.

```python
def greet(name):
    print(f"Hello, {name}!")

greet("Maria")
```

## Returning values

`return` sends a value back. I use this when I want to store the result or use it in another calculation.

```python
def add_tax(price, tax_rate):
    tax = price * tax_rate
    return price + tax

final_price = add_tax(16, 0.1875)
round(final_price, 2)
```

## Default values

A parameter can have a default value. This is useful when most calls use the same value.

```python
def emphasize(text, marks="!!!"):
    return text + marks

emphasize("Python", ""), emphasize("Python")
```

## Mini exercise

Write a function that receives a list of numbers and returns the average.

```python
def average(numbers):
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)

average([7, 8, 9])
```

Note to myself: functions should usually do one clear thing. If a function gets hard to name, maybe it is doing too much.
