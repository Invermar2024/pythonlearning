# 02 - Conditionals and Loops

This notebook is about making decisions and repeating actions. These two ideas show up in almost every real program.

## Comparisons

A comparison gives me `True` or `False`. I still mix up `=` and `==` sometimes: `=` assigns a value, `==` compares two values.

```python
age = 20

age >= 18, age == 18, age != 18
```

## If statements

`if` lets the program choose what to do. The indentation matters because it shows which lines belong to the condition.

```python
if age >= 18:
    print("You can enter.")
else:
    print("You are too young.")
```

## For loops

A `for` loop is useful when I know the collection I want to go through.

```python
topics = ["variables", "lists", "loops"]

for topic in topics:
    print(f"Practice: {topic}")
```

## While loops

`while` loops keep going while a condition is true. I need to make sure something changes inside the loop, or it can run forever.

```python
counter = 3

while counter > 0:
    print(f"Countdown: {counter}")
    counter -= 1
```

## Mini exercise

Loop through a list of grades and print whether each one is passing. I chose `60` as the passing score.

```python
grades = [75, 42, 88, 59]

for grade in grades:
    if grade >= 60:
        print(f"{grade}: pass")
    else:
        print(f"{grade}: review again")
```
