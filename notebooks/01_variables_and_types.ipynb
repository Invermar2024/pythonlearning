## Variables

A variable is a name that points to a value. The value can be text, a number, a list, or many other things.

```python
name = "Nicolas"
age = 28

print(f"My name is {name} and I am {age} years old.")
```

Something I noticed: Python does not need me to write the type before the variable. It figures it out from the value.

```python
type(name)
```

```python
type(age)
```

## Numbers

I want to remember the difference between integers and decimals:

- `int`: whole number
- `float`: number with decimals

```python
temperature_morning = 21
temperature_afternoon = 21.5

temperature_morning, temperature_afternoon, type(temperature_morning), type(temperature_afternoon)
```

## Strings

Strings are text. I use them for names, messages, file paths, and anything that should be treated like words instead of math.

```python
sentence = "python feels easier when i test small pieces first."

print(sentence)
print(sentence.upper())
print(sentence.capitalize())
```

I made a small note here: methods like `.upper()` return a changed version of the string, but the original variable is still the same unless I assign it again.

```python
sentence
```

## Lists

A list keeps multiple values in order. This feels useful when I need to collect items and loop over them later.

```python
topics = ["variables", "strings", "lists"]
topics.append("dictionaries")

topics
```

```python
for index, topic in enumerate(topics, start=1):
    print(f"{index}. {topic}")
```

`enumerate()` is one of those things I want to remember. It gives me the position and the value at the same time.

## Dictionaries

A dictionary stores values with names called keys. This is easier to read than remembering list positions.

```python
student = {
    "name": "Ana",
    "role": "student",
    "progress": 3,
}

student
```

```python
print(f"{student['name']} has completed {student['progress']} lessons.")
```
